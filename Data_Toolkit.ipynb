**Ques2 :- How does broadcasting work in NumPy**
- Broadcasting in NumPy is a powerful feature that allows operations between arrays of different shapes without explicitly reshaping or copying data. It works by stretching the smaller array across the larger one so they have compatible shapes for element-wise operations.
- Broadcasting Rules
- For two arrays to be broadcastable, their dimensions must satisfy the following:
- If the two arrays differ in number of dimensions, the smaller array is padded with ones on the left side.
- Dimensions are compatible when:
- They are equal, or
- One of them is 1
- If these conditions are met, NumPy stretches (broadcasts) the smaller array along the mismatched dimension(s).

**Ques8 :- What does the term “vectorized operation” mean in NumPy ?**
- In NumPy, a vectorized operation means performing operations on entire arrays (vectors, matrices) at once without using explicit loops.
- It allows NumPy to apply a function or arithmetic operation to all elements simultaneously, using optimized C-based implementations under the hood.
- **Advantages of Vectorized Operations:-**
  - Faster Execution: Uses optimized C code instead of Python loops.
  - Cleaner Code: Fewer lines, easier to read and maintain.
  - Memory Efficient: Operations happen in bulk, minimizing overhead.
  - Consistency: Automatically handles element-wise operations.

```python
# Without vectorization (using Python loops):
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]
result = []

for i in range(len(list1)):
    result.append(list1[i] + list2[i])

print(result)
# Output: [6, 8, 10, 12]

#With vectorization (using NumPy):

import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
result = a + b
print(result)
# Output: [6 8 10 12]
```

**Ques13 :- Why is handling missing data important in Pandas**
- Handling missing data in Pandas is extremely important because missing or null values (NaN) can distort analysis, cause errors, or lead to wrong conclusions if not managed properly.
- **Reasons Why Handling Missing Data Is Important:-**
 - **1. Ensures accurate analysis:**
   - Missing values can affect calculations like mean, sum, or correlations, leading to incorrect results.
- **2. Prevents model errors:**
   - Machine learning algorithms and statistical models usually cannot handle NaN values, so cleaning is essential before training
 - **3. Maintains data integrity:**
   - Missing data might indicate issues in data collection or entry, so identifying them helps maintain dataset quality.

```python
import pandas as pd

data = {'Name': ['Amit', 'Riya', 'Karan', 'Priya'],
        'Age': [25, None, 28, None]}
df = pd.DataFrame(data)

print("Before Handling:\n", df)

# Fill missing values with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

print("\nAfter Handling:\n", df)
```

**Ques17 :- Explain the difference between apply() and map() in Pandas**
- **1. map()**
- **Purpose:** Applies a function to each element of a Series individually.
- **Works on:** Series only (cannot be used directly on DataFrame).
- **Input:** A function, dictionary, or Series.
- **Returns:** A new Series with transformed values.

- **2. apply()**
- **Purpose:** Applies a function along any axis of a DataFrame or to each element of a Series.
- **Works on:** Series or DataFrame.
- **Input:** A function (can be more complex than map).
- **Returns:**
  - Series when applied to Series
  - DataFrame or Series when applied along DataFrame axis

```python
# map()

import pandas as pd

s = pd.Series([1, 2, 3, 4])
squared = s.map(lambda x: x**2)
print(squared)

print("---------------------------------------------")
# apply()

df = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6]})
df_sum = df.apply(sum, axis=0)  # Sum of each column
print(df_sum)
```

**Ques21 :- Why is NumPy’s array slicing faster than Python’s list slicing**
- NumPy’s array slicing is faster than Python’s list slicing because of the way NumPy arrays are stored and managed in memory, and how operations are implemented at a low level.
- **Key Reasons for the Speed Advantage:**
- **1. Contiguous Memory Storage**
  - NumPy arrays store data in contiguous blocks of memory (like C arrays), allowing fast access to elements.
  - Python lists store pointers to objects scattered in memory, making access slower.
- **2. Vectorized Operations**
  - NumPy operations, including slicing, are implemented in compiled C code internally, avoiding Python-level loops.
  - Python lists require interpreted Python loops for many operations.
- **3. Views Instead of Copies**
  - Slicing a NumPy array returns a view (reference to the original data) rather than creating a new copy.

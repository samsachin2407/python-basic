**Ques1: What is the difference between a function and a method in Python?**
**Functions :-**
   -  A function is a block of code that is defined outside of a class and is not tied to any object.
   - Defined using: def keyword
   - Called like: function_name(arguments)
   
**Method :-**
   - A method is a function that is defined inside a class and is associated with an object (instance) of that class.
   - Always has self as the first parameter (for instance methods)
   - Called like: object.method_name(arguments)


```python
# defining of function
def greet(name):
    return (f"Hello, {name}!")

print(greet("Sachin"))

# defining of method
class Greeter:
    def greet(self, name):
        return (f"Hello, {name}!")

g = Greeter()
print(g.greet("Sachin"))
```

**Ques2:-  Explain the concept of function arguments and parameters in Python**
-   **Parameter :-** A variable in a function definition. It's like a placeholder.
-  **Argument :-** The actual value passed to the function when you call it.

```python
#  Arguments and Parameter
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Sachin")     # "Sachin" is an argument


#  Positional Arguments
def add(a, b):
    return a + b

add(3, 5)  # a = 3, b = 5

#  Keyword Arguments
def add(a, b):
    return a + b

add(b=5, a=3)  # Same result, order doesn't matter

#  Default Arguments
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()        # Uses default: "Hello, Guest!"
greet("Sachin")  # Overrides default: "Hello, Zara!"


#  Variable-length Arguments
def total(*numbers):
    return sum(numbers)

total(1, 2, 3, 4)  # Returns 10

#  **kwargs – for keyword arguments
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Sachin", age=30)

```

**Ques3:- What are the different ways to define and call a function in Python?**
-  In Python, functions are blocks of reusable code that perform a specific task. There are multiple ways to define and call functions, depending on your needs. Below is a breakdown of the different ways to define and call functions in Python
- **1. Standard Function Definition (def)**


```python
def greet(name):               # Defining the function
    return f"Hello, {name}!"

print(greet("Sachin"))         # Calling the function
```

```python
def shout(text):                             # Defining the function
    return text.upper()

def whisper(text):                           # Defining the function
    return text.lower()

def speak(style_func, message):
    return style_func(message)
print(speak(shout, "hello"))                 # Calling the function
print(speak(whisper, "HELLO"))               # Calling the function
```

**Ques5 :-  What are iterators in Python and how do they differ from iterables**
- **iterators :-**
  - An iterable is any object in Python that can be looped over (i.e., you can use it in a for loop).
   - **Common examples:**  list, tuple, string, set, dict, range
   - It implements the __iter__() method, which returns an iterator.

- **Iterator :-**
  - An iterator is an object that represents a stream of data — it produces items one at a time when you call next() on it.
  - **It implements both:-**
      - __iter__() — returns the iterator itself.
      - __next__() — returns the next item (or raises StopIteration if exhausted).

```python
nums = [1, 2, 3]  # This is an iterable
it = iter(nums)   # This creates an iterator from the iterable

print(next(it))  # Output: 1
print(next(it))  # Output: 2
print(next(it))  # Output: 3
```

**Ques6 : Explain the concept of generators in Python and how they are defined**
- A generator is a special type of iterable that yields values one at a time using the yield keyword instead of returning them all at once.
- **Key Characteristics :-**
  - Defined like a normal function, but uses yield instead of return.
  - Produces values lazily — only when requested via iteration.
  - Maintains internal state between calls.
  - Useful for streaming data or generating sequences on the fly.

- **Define a Generator :-**
   - You define a generator function using the def keyword, just like a normal function, but instead of return, you use yield.

   

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
counter = count_up_to(5)

for num in counter:
    print(num)


```

**Ques7 :- What are the advantages of using generators over regular functions?**
- Generators offer several key advantages over regular functions (or even regular iterables like lists), especially in contexts where performance and memory efficiency are important.
- **Advantages of Using Generators**
- **1. Memory Efficiency :-**
  - Generators don’t store the entire result in memory. They yield one value at a time.
  - Great for large datasets or infinite sequences.

```python
def count_up_to(n):
    for i in range(n):
        yield i
counter = count_up_to(5)
for num in counter:
    print(num)

```

**2. Lazy Evaluation:-**
- Generators generate values on the fly.
- Execution is paused at each yield and resumed when needed.


```python
def read_lines(filename):
    with open(filename) as f:
        for line in f:
            yield line
```

**Ques8 :- What is a lambda function in Python and when is it typically used?**
- A lambda function in Python is a small anonymous function defined using the `lambda` keyword. It can take any number of arguments, but can only have a single expression (no statements, loops, or complex logic).

- **When Is lambda Typically Used:-**
- Lambda functions are commonly used in places where you need a short function temporarily, often as an argument to higher-order functions.
- **1. With map() :-**
   - Applies a function to all items in an iterable.
- **2. With filter() :-**
  - Filters items based on a condition.
- **3. With reduce() (from functools) :-**
   - Used to perform cumulative computation.
   

```python
# Example
add = lambda x, y: x + y
print(add(2, 3))  # Output: 5

# Example with map()
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)  # [1, 4, 9, 16]

# Example with filter()

nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # [2, 4]



# Example with reduce()
from functools import reduce
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)  # 24


```

**Ques9 :-  Explain the purpose and usage of the `map()` function in Python.**
-  The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new iterator with the results.
- **Purpose of map() :-**
  - To transform data from one form to another.
  - To avoid explicit for loops when applying the same function to multiple items.
  - To write more concise and functional-style Python code.

```python
a = [1, 2, 3]
b = [4, 5, 6]
sums = list(map(lambda x, y: x + y, a, b))
print(sums)  # [5, 7, 9]
```

**Ques5 : Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci_sequence(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b
a= fibonacci_sequence(5)
print(a)
```

**Que6: Write a generator function in Python that yields the powers of 2 up to a given exponent**

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent

def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in power_of_two(5):
    print(power)
```

**Ques7 : Implement a generator function that reads a file line by line and yields each line as a string**

```python
# Implement a generator function that reads a file line by line and yields each line as a string

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
read_file_line_by_line('file.txt')
```
